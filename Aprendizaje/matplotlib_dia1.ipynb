```python
from sklearn.datasets import fetch_california_housing
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


x = 'MedInc'
y = 'MedHouseVal'

data = fetch_california_housing(as_frame=True)
df = data.frame
df.head()

plt.title('MedInc vs MedHouseVal')
plt.xlabel('MedInc')
plt.ylabel('MedHouseVal')

plt.scatter(df[x], df[y], alpha=0.1)
plt.show()

```

```python
from sklearn.datasets import fetch_california_housing
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


a = 'MedInc'
b = 'MedHouseVal'
c = 'HouseAge'
d = 'AveRooms'

data = fetch_california_housing(as_frame=True)
df = data.frame
df.head()

plt.subplot(1, 3, 1)
plt.title('MedInc vs MedHouseVal')
plt.xlabel('MedInc')
plt.ylabel('MedHouseVal')
plt.scatter(df[a], df[b], alpha=0.1)

plt.subplot(1, 3, 2)
plt.title('HouseAge vs MedHouseVal')
plt.xlabel('HouseAge')
plt.ylabel('MedHouseVal')
plt.scatter(df[c], df[b], alpha=0.1)

plt.subplot(1, 3, 3)
plt.title('AveRooms vs MedHouseVal')
plt.xlabel('AveRooms')
plt.ylabel('MedHouseVal')
plt.scatter(df[d], df[b], alpha=0.1)

plt.tight_layout()

plt.show()



```
