```python
from nltk import Text
from textblob import TextBlob

#Se agrega el texto como variable
texto_1 = TextBlob("I love this resttaurant, the food was amazing.")
print(texto_1.sentiment)

texto_2 = TextBlob("I hate this restaurant, the food was bad.")
print(texto_2.sentiment)

texto_3 = TextBlob("The food was regular.")
print(texto_3.sentiment)



#clasificador
def clasificar(texto):
    blob = TextBlob(texto)
    polaridad = blob.sentiment.polarity

    if polaridad > 0.1:
      return "positivo"
    elif polaridad < -0.1:
      return "negativo"
    else:
      return "neutral"

#lista de reseñas

reseñas = ["I love this restaurant, the was delicious", "I hate this place", "the food was regular", "this place is amazing, i love your food",
           "this place is awful"]
for reseña in reseñas:
  etiqueta = clasificar(reseña)
  print(f"{reseña} → {etiqueta}")
```
