```python
from nltk import Text
from textblob import TextBlob
import matplotlib.pyplot as plt

#Se agrega el texto como variable
texto_1 = TextBlob("I love this resttaurant, the food was amazing.")
print(texto_1.sentiment)

texto_2 = TextBlob("I hate this restaurant, the food was bad.")
print(texto_2.sentiment)

texto_3 = TextBlob("The food was regular.")
print(texto_3.sentiment)



#clasificador
def clasificar(texto):
    blob = TextBlob(texto)
    polaridad = blob.sentiment.polarity

    if polaridad > 0.1:
      return "positivo"
    elif polaridad < -0.1:
      return "negativo"
    else:
      return "neutral"

#lista de reseñas

reseñas = ["I love this restaurant, the was delicious", "I hate this place", "the food was regular", "this place is amazing, i love your food",
           "this place is awful"]
for reseña in reseñas:
  etiqueta = clasificar(reseña)
  print(f"{reseña} → {etiqueta}")



  import pandas as pd

reseñas_data = {
    "texto": [
        "This product is absolutely amazing, I love it!",
        "Terrible experience, worst purchase ever.",
        "The item arrived on time.",
        "Great quality, highly recommend to everyone!",
        "Complete waste of money, very disappointed.",
        "It works as expected, nothing special.",
        "Fantastic product, exceeded my expectations!",
        "Awful quality, broke after one day.",
        "Decent product for the price.",
        "I am so happy with this purchase, incredible!",
        "Do not buy this, total garbage.",
        "Average product, does the job.",
        "Best purchase I have made this year!",
        "Very poor quality, not worth it at all.",
        "It is okay, nothing outstanding.",
        "Absolutely love it, perfect in every way!",
        "Horrible product, completely useless.",
        "Neutral experience, met basic expectations.",
        "Outstanding quality, will buy again!",
        "Disappointed with the quality, expected better."
    ]
}

df = pd.DataFrame(reseñas_data)
print(df.shape)
df.head()


df['sentimiento'] = df['texto'].apply(clasificar)
df.head()


#graficas de sentimiento

grafica_sentimiento = df['sentimiento'].value_counts()

grafica_sentimiento.plot(kind='bar', title='Distribución de Sentimientos')
plt.xlabel('Sentimiento')
plt.ylabel('Frecuencia')
plt.show()


```
