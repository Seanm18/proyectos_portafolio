```python
import numpy as np
#reto 5 - mas parecido a ML
#Formula  -  precios_normalizados = (precios - minimo) / (maximo - minimo)

precios = np.array([150000, 200000, 350000, 480000, 95000, 620000])

precios_normalizados = (precios - precios.min()) / (precios.max() - precios.min())

print(precios_normalizados)
```
